```python
import os
import time
import getpass
import pandas as pd
import chromadb #vector database which runs locally in the server
import google.generativeai as genai
import google.api_core as exceptions

if "GEMENI_API_KEY" not in os.environ:
  os.environ["GEMENI_API_KEY"]=getpass.getpass('provide your api key')
```

```python
# creating vector databses and search with chromadb
client=chromadb.Client()
file=client.get_or_create_collection(name='medical_guidlines')
data=pd.read_csv('medquad.csv')
```

```python
doccuments=[]
metadatas=[]
ids=[]
```

```python
url_mapping = {
    "NIHSeniorHealth": "nihseniorhealth.gov",
    "NIDDK": "niddk.nih.gov",
    "GARD": "rarediseases.info.nih.gov",
    "MedlinePlus": "medlineplus.gov",
    "Cancer.gov": "cancer.gov",
    "CDC": "cdc.gov",
    "NHLBI": "nhlbi.nih.gov"
}
```

```python
print(f"in process {len(data)} are the medical records . pls wait......")
```

```python
#creating dataset
#takes approx 47 minutes
data=data.dropna(subset='answer') #cleaning the dataset where it is missing
for index, row in data.iterrows():
  text=f"question: {row['question']}     answer: {row['answer']}"
  doccuments.append(text)
  source=str(row['source']) if 'source' in data.columns else "MedQuAD Database"
  source_url = url_mapping.get(source, "kaggle.com/datasets/gpreda/medquad")
  metadatas.append({
      'source':source,
      'url':source_url
  })
  ids.append(f"medquad_{index}")
batch_size = 5000
print("Loading data into ChromaDB in batches...")

for i in range(0,len(doccuments),batch_size):
  print('loding batch size ',i, 'to',i+batch_size)
  file.upsert(
        documents=doccuments[i : i + batch_size],
        metadatas=metadatas[i : i + batch_size],
        ids=ids[i : i + batch_size]
    )
print('now the model knows the facts')
```
